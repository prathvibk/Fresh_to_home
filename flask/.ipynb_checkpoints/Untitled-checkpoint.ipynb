```python
from ftplib import FTP

# Your DriveHQ credentials
USERNAME = "BHOOMIKASURESH33"  # Replace with your DriveHQ username
PASSWORD = "*India123"         # Replace with your DriveHQ password

# FTP server details
FTP_SERVER = "ftp.drivehq.com"
REMOTE_PATH = "/static/upload"  # Path on the DriveHQ server
LOCAL_FILE = "DO_Files/file_7.txt"  # Path to the local file to upload

def upload_file_ftp(server, username, password, local_file, remote_path):
    try:
        # Connect to the FTP server
        ftp = FTP(server)
        ftp.login(username, password)
        print("Connected to the FTP server successfully!")

        # Change to the desired directory on the server
        ftp.cwd(remote_path)

        # Open the local file in binary mode and upload
        with open(local_file, 'rb') as file:
            ftp.storbinary(f"STOR {local_file.split('/')[-1]}", file)
        
        print("File uploaded successfully!")
        ftp.quit()
    except Exception as e:
        print("An error occurred:", e)

# Upload the file
upload_file_ftp(FTP_SERVER, USERNAME, PASSWORD, LOCAL_FILE, REMOTE_PATH)
```

```python
from ftplib import FTP

# Your DriveHQ credentials
USERNAME = "BHOOMIKASURESH33"  # Replace with your DriveHQ username
PASSWORD = "*India123"         # Replace with your DriveHQ password

# FTP server details
FTP_SERVER = "ftp.drivehq.com"
REMOTE_PATH = "/static/heart_disease_user_input/"  # Path on the DriveHQ server
LOCAL_FILE = "user_input/0_test_71.csv"  # Path to the local file to upload

def upload_file_ftp(server, username, password, local_file, remote_path):
    try:
        # Connect to the FTP server
        ftp = FTP(server)
        ftp.login(username, password)
        print("Connected to the FTP server successfully!")

        # Enable Passive Mode (default)
        ftp.set_pasv(True)

        # Change to the desired directory on the server
        ftp.cwd(remote_path)

        # Open the local file in binary mode and upload
        with open(local_file, 'rb') as file:
            ftp.storbinary(f"STOR {local_file.split('/')[-1]}", file)
        
        print("File uploaded successfully!")
        ftp.quit()
    except Exception as e:
        print("An error occurred:", e)

# Upload the file
upload_file_ftp(FTP_SERVER, USERNAME, PASSWORD, LOCAL_FILE, REMOTE_PATH)
```
